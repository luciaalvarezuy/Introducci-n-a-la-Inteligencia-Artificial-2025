2. Visualización de Datos
Una buena manera de hacer que los estudiantes entiendan los patrones es mediante gráficos.

2.1. Histograma de la Renta Disponible
Este gráfico ayuda a ver cómo se distribuye la renta entre todas las secciones censales.

```python
import matplotlib.pyplot as plt

plt.hist(df_renta["Import_Euros"], bins=30, edgecolor="black")
plt.xlabel("Renta Disponible (€)")
plt.ylabel("Frecuencia")
plt.title("Distribución de la Renta Disponible en Barcelona")
plt.show()
```

2.2. Gráfico de Caja (Boxplot) de la Renta por Distrito
Sirve para ver la dispersión y posibles valores atípicos.

```python
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.boxplot(x="Nom_Districte", y="Import_Euros", data=df_renta)
plt.xticks(rotation=90)
plt.title("Distribución de la Renta por Distrito")
plt.show()
```

2.3.  Gráfico de Barras (Comparación de Distritos)
Para comparar la renta promedio en cada distrito.

```python

plt.figure(figsize=(10,6))
sns.barplot(x="Nom_Districte", y=df_renta["Import_Euros"] / 1000, data=df_renta, estimator=sum, ci=None)
plt.xticks(rotation=45)
plt.title("Suma de Renta por Distrito (en miles de euros)")
plt.xlabel("Distrito")
plt.ylabel("Total en Miles de Euros")
plt.show()

```
